```python
## Calculating Aggregate Functions III
```

```python
df = lambda x: np.percentile(x, 25)
```

```python
cheap_category = orders.head(10).groupby('category').amount.apply(df).reset_index()
```

```python
cheap_category
```

```python
cheap_category = orders.head(10).groupby('category').count().reset_index()
```

```python
cheap_category
```

```python
## Calculating Aggregate Functions IV {Count}
```

```python
orders_counts = orders.groupby(['category','subcategory']).ID.count().reset_index()
```

```python
orders_counts
```

```python
##  Pivot Tables {dikelompokkan gitu/sama aja kayak orders_counts}
```

```python
orders_counts_pivot = orders_counts.pivot(
  columns='category',
  index='subcategory',
  values='ID'
).reset_index()
```
